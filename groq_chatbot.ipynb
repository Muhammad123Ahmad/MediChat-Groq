```python
import gradio as gr
from groq import Groq

# Read API key from txt file
with open("groq_api_key.txt", "r") as f:
    API_KEY = f.read().strip()

# Initialize the Groq client
client = Groq(api_key=API_KEY)

# Chat history will be kept here
chat_history = [
    {
        "role": "system",
        "content": (
            "You are a helpful and knowledgeable medical assistant. "
            "You provide accurate, safe, and clear information about "
            "symptoms, conditions, treatments, medications, and healthy living. "
            "Always encourage users to consult a licensed healthcare professional "
            "for diagnosis or urgent medical concerns."
        ),
    }
]

def medical_chatbot(user_input, history):
    # Add user message to history
    chat_history.append({"role": "user", "content": user_input})
    
    # Create completion
    completion = client.chat.completions.create(
        model="llama3-8b-8192",
        messages=chat_history,
        temperature=0.7,
        max_completion_tokens=512,
        top_p=1,
    )
    
    # Get assistant reply
    reply = completion.choices[0].message.content
    
    # Save assistant reply in history
    chat_history.append({"role": "assistant", "content": reply})
    
    return reply

# Gradio UI
with gr.Blocks() as demo:
    gr.Markdown("## 🩺 Medical Chatbot\nAsk me anything about health, symptoms, or wellness. (Not a substitute for professional medical advice!)")
    
    chatbot = gr.Chatbot()
    msg = gr.Textbox(placeholder="Type your medical question here...")
    clear = gr.Button("Clear Chat")
    
    def respond(message, chat_history_ui):
        reply = medical_chatbot(message, chat_history_ui)
        chat_history_ui.append((message, reply))
        return "", chat_history_ui

    msg.submit(respond, [msg, chatbot], [msg, chatbot])
    clear.click(lambda: ([], []), None, [chatbot, msg])

# Launch Gradio app with public link
demo.launch(share=True)
```
